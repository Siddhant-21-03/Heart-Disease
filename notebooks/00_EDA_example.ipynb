from pathlib import Path
import sys
import pandas as pd
# Robust project-root detection: walk upward until a folder containing 'src' is found
def find_project_root(start: Path = Path.cwd()):
    p = start.resolve()
    for ancestor in [p] + list(p.parents):
        if (ancestor / 'src').is_dir():
            return ancestor
    return p
